```python
#Herramientas procesamiento de texto
#Expresiones regulares
regex_special = "/(\W)"
regex_numbers = "/(\d)"

#ListasFiltro
colores = ['negro', 'blanco', 'amarillo', 'rojo', 'azul', 'verde', 'naranja', 'cafe', 'gris', 'plata', 'beige', 'rosa']
meses = ['enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre', 'octubre', 'noviembre', 
         'diciembre','sep', 'oct', 'nov', 'dic']
unnecesary = ['con', 'un', 'unos', 'alta', 'baja', 'marca', 'serie', 'modelo', 'frente', 'altura', 'ancho', 
              'las', 'los', 'para', 'con', 'por', 'color',]
```

```python
#Filtros selectores por categorías
selector_equipo_computo = ['cpu', 'gabinete', 'computador','teclado', 'monitor', 'intel', 'amd', 'ssd', 'hdd', 'raton', 'mouse',
                   'impresora', 'nobreak', 'regulador','bocina', 'audifono', 'microfono', 'scaner', 'escaner', 'pantalla',
                   'proyector', 'television', 'antena','telefono', 'dvd', 'reproductor', 'vhs', 'copiadora', 'laptop', 
                           'casetera', 'cassetera', 'vga', 'usb', 'discoduro','ipad', 'radio', 'switch', 'utp', 'probador',
                          'web', 'windows', 'mac', 'gabinete', 'memoria', 'mb', 'gb', 'tb', 'ddr', 'pila', 'bateria',
                          'camara', 'contacto', 'sony', 'philips', 'hp', 'toshiba', 'motorola', 'dell', 'asus']

selector_transporte = ['camioneta', 'automovil', 'carro', 'lancha', 'transporte', 'pasajeros', 'placas', 'toyota', 
                       'mercedes', 'nissan', 'dodge', 'honda', 'ford']

selector_material_oficina = ['pluma', 'carpeta', 'cuaderno', 'marcatexto', 'zacapuntas', 'marcador']

selector_marcas = ['sony', 'philips', 'hp', 'toshiba', 'motorola', 'dell', 'asus', 'toyota', 'mercedes', 'nissan', 
                   'dodge', 'honda', 'ford', 'trupper', 'dewalt', 'apple', 'panasonic', 'acer', 'intel', 'amd']

selector_libros = ['libro', 'isnb', 'editorial', 'publicacion', 'autor']

selector_escolar = ['butaca', 'pizarron', 'pintarron']

selector_herramienta = ['martillo', 'desarmador', 'pinzas', 'llave', 'alicates', 'flexometro', 
                        'taladro', 'alicates', 'lampara', 'pintura', 'ml']

selector_prot_civil = ['extintor', 'chaleco', 'botiquin', 'lampara', 'emergencia']

selector_material = ['madera', 'mdf','aluminio', 'vinil', 'tela', 'pvc', 'metal', 'cobre', 'aluminio','piel', 
                     'vidrio', 'cristal', 'caoba']

selector_mobiliario = ['escritorio', 'anaquel', 'silla', 'sofa', 'sillon', 'oficina', 'respaldo', 'pizarron', 'pizarra',
                       'portafolio', 'archivero', 'cajon', 'armario', 'lampara', 'mesa', 'cocina', 'estufa', 'comedor'
                      'alacena', 'estante', 'ejecutivo', 'ejecutiva', 'escalera', 'locker', 'cesto', 'bote', 'trituradora']

selector_cubiertos = ['vaso', 'copa', 'taza', 'termo', 'cuchillo', 'tenedor', 'cuchara']
```

```python
#Procesamiento de texto
#Se convierte la columna en una lista
bien_list = df['Bien'].tolist()

for i in range(0, len(bien_list)):
    #Uso de minusculas
    bien_list[i] = str(bien_list[i]).lower()
    
    #Reemplazo de acentos
    transformed_string = str(bien_list[i]).replace("á","a")
    bien_list[i] = transformed_string
    transformed_string = str(bien_list[i]).replace("é","e")
    bien_list[i] = transformed_string
    transformed_string = str(bien_list[i]).replace("í","i")
    bien_list[i] = transformed_string
    transformed_string = str(bien_list[i]).replace("ó","o")
    bien_list[i] = transformed_string
    transformed_string = str(bien_list[i]).replace("ú","u")
    bien_list[i] = transformed_string
    
    #Eliminar colores
    for j in range(0, len(colores)):
        transformed_string = str(bien_list[i]).replace(colores[j],"")
        bien_list[i] = transformed_string
    
    #Eliminar meses
    for j in range(0, len(meses)):
        transformed_string = str(bien_list[i]).replace(meses[j],"")
        bien_list[i] = transformed_string
        
    for j in range(0, len(unnecesary)):
        transformed_string = str(bien_list[i]).replace(unnecesary[j],"")
        bien_list[i] = transformed_string
        
    #Eliminar caracteres especiales y espacios
    bien_list[i] = re.sub(r'\W','', str(bien_list[i]))
    
    #Eliminar numeros o digitos
    bien_list[i] = re.sub(r'\d', '', str(bien_list[i]))
    
print(bien_list)
```

```python
#Clasificacion

temp_bien = df['Bien'].tolist()
temp_apartado = df['Apartado'].tolist()

finalDF = pd.DataFrame(columns = ['Bien','Apartado','Items','Tipo','Subtipo1','Subtipo2','Subtipo3'])

for i in range (len(bien_list)):
    sel_comp = 0
    sel_trans = 0
    sel_mat_of = 0
    sel_marca = 0
    sel_libro = 0
    sel_escolar = 0
    sel_herr = 0
    sel_prot = 0
    sel_mat = 0
    sel_mobi = 0
    sel_cubi = 0
    cad_sel = ""
    
    dictTemp = {}
    dictPuntos = {}
    
    #dictF["Bien"] = temp_bien[i]
    dictTemp['Bien'] = temp_bien[i]
    dictTemp['Apartado'] = temp_apartado[i]
    
    for j in range(len(selector_equipo_computo)):
        if selector_equipo_computo[j] in bien_list[i]:
            sel_comp = sel_comp + 1
            if(sel_comp != 0):
                dictPuntos["Computo"] = sel_comp
                cad_sel = cad_sel + str(selector_equipo_computo[j]) + ","
    
    for j in range(len(selector_transporte)):
        if selector_transporte[j] in bien_list[i]:
            sel_trans = sel_trans + 1
            if(sel_trans != 0):
                dictPuntos["Transporte"] = sel_trans
                cad_sel = cad_sel + str(selector_transporte[j]) + ","
            
    for j in range(len(selector_material_oficina)):
        if selector_material_oficina[j] in bien_list[i]:
            sel_mat_of = sel_mat_of + 1
            if(sel_mat_of != 0):
                dictPuntos["MaterialOficina"] = sel_mat_of
                cad_sel = cad_sel + str(selector_material_oficina[j]) + ","
            
    for j in range(len(selector_marcas)):
        if selector_marcas[j] in bien_list[i]:
            sel_marca = sel_marca + 1
            if(sel_marca != 0):
                dictPuntos["Marca"] = sel_marca
                cad_sel = cad_sel + str(selector_marcas[j]) + ","
    
    for j in range(len(selector_libros)):
        if selector_libros[j] in bien_list[i]:
            sel_libro = sel_libro + 1
            if(sel_libro != 0):
                dictPuntos["Libro"] = sel_libro
                cad_sel = cad_sel = str(selector_libros[j]) + ","
            
    for j in range(len(selector_escolar)):
        if selector_escolar[j] in bien_list[i]:
            sel_escolar = sel_escolar + 1
            if(sel_escolar != 0):
                dictPuntos["Escolar"] = sel_escolar
                cad_sel =  cad_sel + str(selector_escolar[j]) + ","
            
    for j in range(len(selector_herramienta)):
        if selector_herramienta[j] in bien_list[i]:
            sel_herr = sel_herr + 1
            if(sel_herr != 0):
                dictPuntos["Herramienta"] = sel_herr
                cad_sel = cad_sel + str(selector_herramienta[j]) + ","
            
    for j in range(len(selector_prot_civil)):
        if selector_prot_civil[j] in bien_list[i]:
            sel_prot = sel_prot + 1
            if(sel_prot != 0):
                dictPuntos["ProteccionCivil"] = sel_prot
                cad_sel = cad_sel + str(selector_prot_civil[j]) + ","
            
    for j in range(len(selector_material)):
        if selector_material[j] in bien_list[i]:
            sel_mat = sel_mat + 1
            if(sel_mat != 0):
                dictPuntos["Material"] = sel_mat
                print("Material ",sel_mat)
                cad_sel = cad_sel + str(selector_material[j]) + ","
    
    for j in range(len(selector_mobiliario)):
        if selector_mobiliario[j] in bien_list[i]:
            sel_mobi = sel_mobi + 1
            if(sel_mobi != 0):
                dictPuntos["Mobiliario"] = sel_mobi
                cad_sel = cad_sel + str(selector_mobiliario[j]) + ","
            
    for j in range(len(selector_cubiertos)):
        if selector_cubiertos[j] in bien_list[i]:
            sel_cubi = sel_cubi + 1
            if(sel_cubi != 0):
                dictPuntos["Cubiertos"] = sel_cubi
                cad_sel = cad_sel + str(selector_cubiertos[j]) + ","
    
    cad_sel = cad_sel.upper()
    cad_sel = cad_sel[:-1]
    #print(temp_bien[i]+"\n", dictPuntos)
    
    if(len(cad_sel)!=0):
        dictTemp["Items"] = cad_sel
    else:
        dictTemp["Items"] = "Indefinido"
    
    # > 1 Coincidencia 
    if (len(dictPuntos) > 1):
        #Ordenar mayor a menor
        sortDict = dict(sorted((dictPuntos).items(), key=operator.itemgetter(1),reverse=True))
        if(len(dictPuntos) == 2):
            key, value = list(sortDict.items())[0]
            dictTemp["Tipo"] = key
            key, value = list(sortDict.items())[1]
            dictTemp["Subtipo1"] = key
            dictTemp["Subtipo2"] = "Indefinido"
            dictTemp["Subtipo3"] = "Indefinido"
        elif(len(dictPuntos) == 3):
            key, value = list(sortDict.items())[0]
            dictTemp["Tipo"] = key
            key, value = list(sortDict.items())[1]
            dictTemp["Subtipo1"] = key
            key, value = list(sortDict.items())[2]
            dictTemp["Subtipo2"] = key
            dictTemp["Subtipo3"] = "Indefinido"
        elif(len(dictPuntos) >= 4):
            key, value = list(sortDict.items())[0]
            dictTemp["Tipo"] = key
            key, value = list(sortDict.items())[1]
            dictTemp["Subtipo1"] = key
            key, value = list(sortDict.items())[2]
            dictTemp["Subtipo2"] = key
            key, value = list(sortDict.items())[3]
            dictTemp["Subtipo2"] = key
        
        #print(sortDict)
    #Sin coincidencias
    elif (len(dictPuntos) == 0):
        dictTemp["Tipo"] = "Indefinido"
        dictTemp["Subtipo1"] = "Indefinido"
        dictTemp["Subtipo2"] = "Indefinido"
        dictTemp["Subtipo3"] = "Indefinido"
    #1 Coincidencia
    else:
        key, value = list(dictPuntos.items())[0]
        dictTemp["Tipo"] = key
        dictTemp["Subtipo1"] = "Indefinido"
        dictTemp["Subtipo2"] = "Indefinido"
        dictTemp["Subtipo3"] = "Indefinido"
        
    #print(bien_list[i], sel_comp, sel_trans, sel_mat_of, sel_marca, sel_libro, sel_escolar, sel_herr, sel_prot, sel_mat, sel_mobi, sel_cubi)
    #print(dictPuntos)
    
    finalDF = finalDF.append(dictTemp, ignore_index=True)
    print(dictTemp)
```
